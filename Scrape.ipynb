```python
def match(title,tcheck):
    r={}
    k=[]
    for dd in title:
        r[similar(dd,tcheck)]=dd
        k.append(similar(dd,tcheck))
    return r[max(k)] if max(k)>0.1 else pd.np.nan
    
```

```python
def similar(a, b):
    f=0
    for x in list(b.split()):
        if x.lower() not in a.lower().replace(' ',''):
            f+=1
    return SequenceMatcher(None, a, b).ratio() if (f<=3) & (f!=(1/2)*len(list(b.split()))) else 0
```
